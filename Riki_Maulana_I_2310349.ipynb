# Library pandas

```python
#Library Pandas#
#digunakan untuk manipulasi dan analisis data

import pandas as pd
data = {"nama":["ikiw", "riki", "maul"],
        "umur":[21, 25, 90],
        "kota":["jakarta", "kajarta", "takarja"]}
data_2 = pd.DataFrame(data)
print(data_2)
```

```python
display(data_2)
```

```python
#menampilkan data dengan umur khusus lebih besar sama dengan 25
display(data_2[data_2.umur >= 25])
```

```python
display(data_2[data_2.nama == "riki"])
```

```python
display(data_2[(data_2.nama == "riki") | (data_2.nama == "maul")])
```

```python
display(data_2[data_2.nama != "maul"])
```

# library matplotlib

```python
#library matplotlib#
#library untuk membuar grafik/plot

import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
y = np.tan(x)
plt.plot(x,y, label = "grafik tan")
plt.legend()
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
```
